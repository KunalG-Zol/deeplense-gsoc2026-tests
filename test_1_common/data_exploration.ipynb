## Checking the dimensions of the images

```python
img_arr = np.load('dataset/dataset/train/vort/3.npy')
```

```python
np.shape(img_arr)
```

```python
img_arr
```

```python
import matplotlib.pyplot as plt
```

```python
gamma_corrected = np.power(img_arr[0] + 1e-8, 0.2)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
im1 = axes[0].imshow(img_arr[0], cmap='magma')
axes[0].set_title('Raw Min-Max Image\n(What the network currently sees)')
axes[0].axis('off')
fig.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)

# Plot 2: The Transformed Image
im2 = axes[1].imshow(gamma_corrected, cmap='magma')
axes[1].set_title(f'Gamma Corrected\n(Revealing the hidden substructure)')
axes[1].axis('off')
fig.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
